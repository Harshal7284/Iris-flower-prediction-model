```python
df = df.drop(columns = ["Id"])
```

```python
df['Species'] = df['Species'].map({'Iris-setosa' : 0 , 'Iris-versicolor' : 1 , 'Iris-virginica' : 2})
```

```python
from sklearn.model_selection import train_test_split
import xgboost as xgb
```

```python
X=df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
y=df[['Species']]
```

```python
X_test,X_train,y_test,y_train = train_test_split(X,y,test_size = 0.2,random_state=400)
```

```python
model = xgb.XGBClassifier(random_state=400)
```

```python
model.fit(X_train, y_train)
```

```python
y_predict = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, confusion_matrix
```

```python
accuracy = accuracy_score(y_test, y_predict)
accuracy
```

```python
recall = recall_score (y_test, y_predict, average = 'macro')
recall
```

```python
precision = precision_score(y_test, y_predict, average= 'macro')
precision
```

```python
f1 = f1_score(y_test, y_predict, average= 'macro')
f1
```

```python
def prediction_flower():
    SepalLengthCm= float(input("Enter the length of sepal= "))
    SepalWidthCm= float(input("Enter the width of sepal = "))
    PetalLengthCm= float(input("Enter the length of petal = "))
    PetalWidthCm= float(input("Enter the width of petal = "))

    flower_data = pd.DataFrame([[SepalLengthCm, SepalWidthCm, PetalLengthCm, PetalWidthCm]], columns = ['SepalLengthCm' , 'SepalWidthCm' , 'PetalLengthCm' , 'PetalWidthCm'],)
    print("---- Result ----")
    prediction = model.predict(flower_data)[0]
    if prediction == 0 :
        print("Flower is Iris-setosa")
    elif prediction == 1 :
        print("Flower is Iris-versicolor")
    else :
        print("Flower is Iris-verginica")
    
```

```python
prediction_flower()
```
